##Step 4: Install ITSMe

This will take about 50 minutes

The main R package used in this practical is ITSMe, which is hosted on GitHub.
It is not one of the pre-installed R packages in Google Colab.
Because ITSMe depends on several other packages, the installation can take some time.
To avoid long waiting times during the practical session, we install the package in advance and store it in your Google Drive. This ensures that once installed, it can be loaded quickly in future sessions without reinstalling.

###Change runtime type to R

First, change the runtime type of the session to R (ITSMe is an R package, so we need to run R code to install it)

Go to Runtime > Change runtime type > Runtime type > choose R > Save

Keep in mind that you will have to change the runtime back to python if you start to run python code again

```python
#Step 4: Install ITSMe
#This will take about 50 minutes

#The main R package used in this practical is ITSMe, which is hosted on GitHub.
#It is not one of the pre-installed R packages in Google Colab.
#Because ITSMe depends on several other packages, the installation can take some time.
#To avoid long waiting times during the practical session, we install the package in advance and store it in your Google Drive. This ensures that once installed, it can be loaded quickly in future sessions without reinstalling.

# ==========================================================
# ITSMe SETUP STEP 1 — Change runtime type to R
# ==========================================================

#First, change the runtime type of the session to R (ITSMe is an R package, so we need to run R code to install it)
#Go to Runtime > Change runtime type > Runtime type > choose R > Save
#Keep in mind that you will have to change the runtime back to python if you start to run python code again

##Configure persistent R library

drive_lib <- "/content/drive/MyDrive/Rlib"

# Create library folder if it doesn't exist
if (!dir.exists(drive_lib)) {
  dir.create(drive_lib, recursive = TRUE)
}

# Tell R to use Drive library FIRST
.libPaths(drive_lib)

cat("Current library paths:\n")
print(.libPaths())


##Install ITSMe

if (!requireNamespace("ITSMe", quietly = TRUE)) {

  cat("\nITSMe not found. Installing (this may take some time the first time)...\n")

  devtools::install_github(
    "lmterryn/ITSMe",
    build_vignettes = FALSE,
  )

  cat("\nInstallation complete!\n")

} else {
  cat("\nITSMe already installed — loading package.\n")
}

library(ITSMe)

cat("\nITSMe successfully loaded!\n")
```
